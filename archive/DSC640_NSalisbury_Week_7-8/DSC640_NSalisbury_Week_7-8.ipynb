```python
fig, ax = plt.subplots()
fig.set_size_inches(6,6)

sns.despine(top=True, right=True)
s = sns.scatterplot(data=crime_rates[(crime_rates['state']!='United States') &
                                (crime_rates['state']!='District of Columbia')], x="murder", y="burglary")
s.set_xlabel('Murder Rate')
s.set_ylabel('Burglary Rate')
s.set_title('Murder Rate vs Burglary Rate')

```

```python
fig, ax = plt.subplots()
fig.set_size_inches(6,6)

sns.despine(top=True, right=True)
s = sns.scatterplot(data=crime_rates[(crime_rates['state']!='United States') &
                                (crime_rates['state']!='District of Columbia')], 
                    x="burglary", 
                    y="motor_vehicle_theft",
                   size=crime_rates['population'],
                   sizes=(20,5000))
ax.get_legend().remove()
s.set_xlabel('Burlary Rate')
s.set_ylabel('Motor Vehicle Rate')
s.set_title('Burglary Rate vs Motor Vehicle Theft Rate')
```
